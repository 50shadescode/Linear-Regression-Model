```python
from sklearn.linear_model import LinearRegression
import numpy as np 
import pandas as pd

# Independent variable (House size in mÂ²)
house_sizes = np.array([50, 75, 100, 125, 150]).reshape(-1,1)

# Dependent variable (House Price in $)
house_prices = np.array([150000, 200000, 250000, 300000, 350000])

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model using the dataset
model.fit(house_sizes, house_prices)

# Predict house prices based on input house sizes
predicted_prices = model.predict(house_sizes)

# Calculate the difference (error)
differences = house_prices - predicted_prices

# Print the differences
print("Differences between actual and predicted prices:", differences)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 3, 5, 6, 8])

# Model training
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Model evaluation
r2 = r2_score(y, y_pred)
print(f"R-squared: {r2}")

# Plotting
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel("X (Independent Variable)")
plt.ylabel("y (Dependent Variable)")
plt.legend()
plt.show()
```
